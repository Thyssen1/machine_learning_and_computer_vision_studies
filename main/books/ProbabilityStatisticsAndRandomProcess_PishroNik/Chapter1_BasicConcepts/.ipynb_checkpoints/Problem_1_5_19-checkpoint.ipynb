## Problem 1.5.19

For the second order equation $AX^{2} + X + B = 0$ we have that the parameters are $a = A, b = 1, c = B$ and the solution is:

$x = \frac{b- \pm \sqrt{b^{2} - 4ac}}{2a}$

All the real solutions are obtained when $b^{2} - 4ac \geq 0$ which leads to $B \leq \frac{1}{4A}$



```python
!pip install numpy
!pip install matplotlib

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the range for A (avoiding zero to prevent division by zero)
A = np.linspace(0.01, 10, 500)
B = 1 / (4 * A)

# Plot B as a function of A
plt.figure(figsize=(10, 6))
plt.plot(A, B, label=r'$B \leq \frac{1}{4A}$', color='blue')
plt.fill_between(A, B, alpha=0.2, color='blue')  # Shading the area under the curve

# Set labels and title
plt.xlabel('A')
plt.ylabel('B')
plt.title(r'Plot of $B \leq \frac{1}{4A}$ as a function of A')
plt.legend()
plt.grid(True)

# Set xlim and ylim to 1
plt.xlim(0, 1)
plt.ylim(0, 1)

plt.show()
```

In the plot above we see that the area of the unit square where real solutions to the equation exist.

The area is calculated as:

$\frac{1}{4} + \int_{\frac{1}{4}}^{1} \frac{1}{4A} dA = \frac{1}{4} + \frac{1}{4} \ln{4}$

Finally, the probability that the equation has real solutions is:

$P(B \leq \frac{1}{4A}) = \frac{\frac{1}{4} + \frac{1}{4} \ln{4}}{1} = \frac{1}{4} + \frac{1}{4} \ln{4}$
